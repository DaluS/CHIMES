## Method Demonstrations

### 1. `get` Method

The `get` method retrieves the current configuration settings. It can return all settings or specific ones based on provided keys.

#### Demonstration:


```python
# Get all configuration settings
all_config = config.get()
print("All Configuration Settings:")
print(all_config)

# Get specific configuration setting 
verb_config = config.get('_VERB')
print("\n_VERB Configuration Setting:")
print(verb_config)
```

### 2. `set_value` Method

The `set_value` method updates the configuration settings. It can update a single setting or multiple settings at once.

#### Demonstration:


```python
# Update a single configuration setting 
config.set_value('_VERB', False)
print("Updated _VERB to False.")

# Verify the update
verb_config = config.get('_VERB')
print("\nUpdated _VERB Configuration Setting:")
print(verb_config)

# Update multiple configuration settings
# Replace with keys and values that exist in your configuration
config.set_value({'_VERB': True, '_SOLVER': 'custom_solver'})
print("\nUpdated multiple settings: _VERB and _SOLVER.")

# Verify the updates
multi_config = config.get(['_VERB', '_SOLVER'])
print("\nUpdated Configuration Settings:")
print(multi_config)
```

### 3. `reset` Method

The `reset` method resets configuration settings to their default values. It can reset all settings or specific ones.

#### Demonstration:


```python
# Reset a specific configuration setting 
config.reset('_VERB')
print("Reset _VERB to its default value.")

# Verify the reset
verb_config = config.get('_VERB')
print("\nReset _VERB Configuration Setting:")
print(verb_config)

# Reset all configuration settings
config.reset()
print("\nReset all configuration settings to defaults.")

# Verify the reset for all settings
all_config = config.get()
print("\nAll Configuration Settings after Reset:")
print(all_config)
```
