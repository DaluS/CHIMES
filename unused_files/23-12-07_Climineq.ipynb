## Green, Brown, Grey capital

Humans are not directly impacting their environment: it's mostly their infrastructure, productive and non-productive that are the cause through emissions and land use. 

By far, most of the emissions are done through energy production, based on fossil fuels, coal or gas. Those infrastructure are what we call the brown capital
This energy can also be produced by another type of technology, with different property, that we will call here green (solar panel, wind turbine...). It is not possible to transform a unit of brown capital into a unit of green capital: once the unit is created, it cannot truly be changed. 
This energy cannot be consumed directly by the society: it need to be transformed into good and services. It is done here by grey capital, which is not emitting itself but requires energy to do so: it can indirectly emit through the energy it needs, depending if it's green or brown. 

The role of the economy is to allocate the right sources of capital for the best output: enough energy to feed the output (grey) sector, and investing into the most interesting technology based on their rentability and economic purposes.
