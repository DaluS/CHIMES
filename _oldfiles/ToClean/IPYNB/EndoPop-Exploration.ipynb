## Sanity check - Goodwin model

```python
### LOADING MODEL
hub_gw=pgm.Hub('Goodwin',preset='default',verb=False)

### INTROSPECTION 
hub_gw.get_summary()
hub_gw.get_fieldsproperties()
hub_gw.get_dataframe(t0=0,t1=0).transpose()
hub_gw.get_Network(params=True)      
groupsoffields = hub_gw.get_dparam_as_reverse_dict(
    crit='units',
    eqtype=['differential', 'statevar',None])
```

```python
### MINIMAL RUN AND PLOT 
hub_gw.run(N=100)
```

```python
hub_gw.plot()#filters_key=('L'),
         #title='basic GK')
```

```python
pgm.plots.phasespace(hub_gw,
    x='employment',
    y='omega',
    color='d'
)
```
