```python
import numpy as np 
import matplotlib.pyplot as plt

epsilon =1.4 
g = 0.03 
r0 = 0.3 
zeta = .001 
Y0 = 1. 

tmax = (1/g)*np.log( 1 + g*(1-r0)/(zeta*Y0))
t=np.linspace(0,tmax,100)
```

```python
nuup = np.exp( g*t*(1-epsilon))
rt = r0 + (zeta*Y0/g)*(np.exp(g*t)-1)
y = - nuup*rt
```

```python
plt.figure()
ax =plt.gca()
ax2 = plt.twinx(ax)
ax. plot(t/tmax,rt,color='r')
ax2.plot(t/tmax,y,color='b')


ax.set_xlabel(r'$\dfrac{t}{t_{end}}$')
ax.set_ylabel(r'$r(t)$',color='red')
ax2.set_ylabel(r'$y$',color='blue')
ax2.spines['left'].set_color('red')
ax2.spines['right'].set_color('blue')
ax .tick_params(axis='y', colors='red')
ax2.tick_params(axis='y', colors='blue')
plt.title(f'$\epsilon ={epsilon} $ \n'+',$ T_{end}='+f'{tmax:.2f}. g= {g:.3f}, r_0 = {r0:.3f},\zeta = {zeta:.3f}, Y_0 = {Y0:.2f}$' )
plt.show()
```

```python
epsilon =1.4 
g = 0.03 
r0 = 0.1
zeta = .1 
Y0 = 0.1

EPSILON = np.linspace(.4,2,17)
import matplotlib as mpl


plt.figure('')

for i,epsilon in enumerate(EPSILON):
    #tmax = (1/g)*np.log( 1 + g*(1-r0)/(zeta*Y0))
    t=np.linspace(0,105,100)

    nuup = np.exp( g*t*(1-epsilon))
    rt = r0 + (zeta*Y0/g)*(np.exp(g*t)-1)
    y = - nuup*np.log(rt)

    plt.plot(t[rt<1],y[rt<1],color = plt.cm.jet(i/len(EPSILON)),label=f'$\epsilon={epsilon:.2f}$')
    armax = np.argmax(y)
    plt.plot(t[armax],y[armax],'*',c='k')
    #plt.plot(t,rt,c='k')
plt.ylabel('y')
plt.xlabel('time')
plt.legend()
plt.title('$ T_{end}='+f'{tmax:.2f}. g= {g:.3f}, r_0 = {r0:.3f},\zeta = {zeta:.3f}, Y_0 = {Y0:.2f}$')
plt.show()
```
